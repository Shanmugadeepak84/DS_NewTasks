```python
#Confidence Interval is a range where we are certain that true value exists.
import numpy as np
import pandas as pd

import scipy.stats
import scipy.stats as stats
#1.Collect the Sample Data:
# initialize data of lists.
data = {'Durability': [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]}
# Create DataFrame
df = pd.DataFrame(data)
print(df)
#2.Calculate mean,standard deviation and sample size
stddeviation = df['Durability'].std(axis=0)
print(f'std deviation for column-Durability is: {stddeviation}')
mean = df['Durability'].mean(axis=0)
print(f'mean for column-Durability is: {mean}')
sample_size=df['Durability'].count()
print(f'count for column-Durability is: {sample_size}')
#Task1-a. Build 99% Confidence Interval Using Sample Standard Deviation
#We use t-distribution when the sample size is less -n<30 and when population standard deviation is unknown.
#Degrees of Freedom (DegOfFree)
DegOfFree=sample_size-1
print(f'count for column-Durability is: {sample_size}')
# create 99% confidence interval
#mean->sample mean
#stddeviation->sample standard deviation
confint_tdist=stats.t.interval(0.99,df=DegOfFree,loc=mean,scale=stddeviation/np.sqrt(sample_size))
print(f'99% Confidence Interval identified using t-distribution when population standard deviation is not known is: {confint_tdist}')
#Task2-Build 99% Confidence Interval Using Known Population Standard Deviation
#We use z-distribution as we know population standard deviation
given_SD_pop=0.2
confint_zdist=stats.norm.interval(0.95,loc=mean,scale=given_SD_pop/np.sqrt(sample_size))
print(f'99% Confidence Interval identified using z-distribution when population standard deviation is known is: {confint_zdist}')
print(confint_zdist)
```
