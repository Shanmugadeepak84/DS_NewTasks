```python
#Hypothesis -Assumption without any experiment,it is made 
#Step1:Formulate Hypothesis-H0,H1 from the perspective of the researcher
  #H1->what the researcher want to prove->Weekly operating costs is increased
    # Mu>W($1,000 + $5X) -ALternate hypothesis 
  #H0->The weekly operating cost according to the cost model is accurate/Weekly operating costs not increased
    # Mu<=W($1,000 + $5X) -Null hypothesis
 # '=' case belongs to null hypothesis
 #We accept or reject only null hypothesis
#Step2: Set a thresold
#alpha=0.05
#Step3:Collect evidence
   #no_of_samples=25,samplemean(xbar)=3050
   #population mean=600 ,population_sd=25
#p-value(right tail test):probability of getting xbar or high
#Step4: Find the p-value coressponding to aplha 
#Z=x-mu/(sigma_pop/sqrt_n)
#3050-4000/(3000/5)
#Step5: Compare p-value and alpha 
  #p>aplha->failed to reject(Accept) Ho (p low nul low)
  #p<aplha ->reject Ho
  
import scipy.stats as stats
import numpy as np

# Define the population mean
total_no_of_samples=600
W=1000+(5*600)
population_mean = W

# Define the sample mean  and standard deviation
sample_mean = 3050
sample_std = 125

# Define the sample size
sample_size = 25

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Define the degrees of freedom
df = sample_size - 1

# Set the significance level (alpha)
alpha = 0.05

# Calculate the critical t-value
critical_t = stats.norm.ppf(1 - alpha)

# Calculate the p-value
p_value = stats.t.sf(abs(t_statistic), df) # survival function (1 - cdf) for one-tailed

# Print the results
print("T-Statistic:", t_statistic)
print("Critical t-value:", critical_t)
print("P-Value:", p_value)

# Decision
#Decision Criteria:

#Reject H₀ if the t-statistic < critical t-value or if p-value < alpha.
#Accept H₀ if the t-statistic > critical t-value or if p-value > alpha.

print('With T-value :')
if t_statistic > critical_t:
   print("""We accept the null hypothesis (H₀)-The weekly operating cost according to the cost model is accurate""")
else:
    print("""We reject the null hypothesis (H₀)-Weekly operating costs is increased""")

print('With P-value :')
if p_value >alpha:
    print("""We accept the null hypothesis (H₀)-The weekly operating cost according to the cost model is accurate""")
else:
    print("""We reject the null hypothesis (H₀)-Weekly operating costs is increased""")

```
