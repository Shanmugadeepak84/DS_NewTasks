```python
#1. State the Hypotheses:
 #H0:Null Hypothesis-there is no dependency between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) 
 #H1:ALternate Hypothesis-there is a dependency between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) 
 
#2. Compute the Chi-Square Statistic:
#Contingency table
import scipy.stats as stats
from scipy.stats import chi2_contingency, chi2
table = [[50, 70], [80, 100], [60, 90], [30, 50], [20, 50]]
alpha = 0.05


stat, p, dof, expected = chi2_contingency(table)
print(f'dof={dof}')
print(f'p-value={p}')
print('Expected Frequencies:')
print(expected)

#Using p-value for decision
# Compare the p-value with alpha
if p <= alpha:
    print("Reject H0 (Null Hypothesis), There is a dependency between the type of smart home device purchased (Smart Thermostats vs. Smart Lights)")
else:
    print("Retain H0 (Null Hypothesis), There is no dependency between the type of smart home device purchased (Smart Thermostats vs. Smart Lights)")
#3.Determine the Critical Value
critical_value = chi2.ppf(q=1-alpha, df=dof)
print(f'Critical value={critical_value}')

# Calculate the Chi-Square Statistic
chi_square = sum(((o-e)**2/e).sum() for o, e in zip(table, expected))
print(f'Chi-square statistic={chi_square}')

#4.Make a Decision Chi-Square statistic with the critical value-Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.
if chi_square >= critical_value:
    print("Reject H0 (Null Hypothesis), There is a dependency between the type of smart home device purchased (Smart Thermostats vs. Smart Lights)")
else:
    print("Retain H0 (Null Hypothesis), There is no dependency between the type of smart home device purchased (Smart Thermostats vs. Smart Lights)")
```
